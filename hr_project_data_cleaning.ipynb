```python
# Phân loại nhân viên: Tạo cột 'Is_Left' (Đã nghỉ = 1, Đang làm = 0)
df_employee['Is_Left'] = df_employee['Exit_Date'].notnull().astype(int)
```

```python
#Tính điểm đánh giá và độ hài lòng của nhân viên trong toàn bộ thời gian làm việc
avg_performance = df_performance.groupby('Employee_Id')[['Manager_Evaluation', 'Employee_Satisfaction']].mean().reset_index()
```

```python
# Gộp bảng Employee và bảng trung bình avg_performance
df_merged = pd.merge(df_employee, avg_performance, on='Employee_Id', how='left')
```

```python
# So sánh lương, điểm đánh giá và độ hài lòng giữa 2 nhóm (Đã việc vs Đang làm việc)
comparison = df_merged.groupby('Is_Left')[['Base_Salary_Annual', 'Manager_Evaluation', 'Employee_Satisfaction']].mean()
```

```python
display(comparison)
```

```python
comparison.index = ['Working', 'Left']
comparison['Base_Salary_Annual'] = comparison['Base_Salary_Annual'].apply(lambda x: f"{x:,.2f}")
comparison['Manager_Evaluation'] = comparison['Manager_Evaluation'].apply(lambda x: f"{x:.2f}")
comparison['Employee_Satisfaction'] = comparison['Employee_Satisfaction'].apply(lambda x: f"{x:.2f}")
```

```python
#Chọn lọc "Gà Nòi"
# 1. Lọc danh sách nhân viên chưa nghỉ việc
active_emp = df_employee[df_employee['Exit_Date'].isnull()]
```

```python
# 2. Tính toán các chỉ số trung bình và tổng giờ tham gia tập huấn từ bảng Performance
talent_stats = df_performance.groupby('Employee_Id').agg({
    'Performance_Rating': 'mean',
    'Engagement_Index': 'mean',
    'Training_Hours': 'sum'
}).reset_index()
```

```python
# 3. Kết nối thông tin nhân viên với bảng thành tích
top_talents = pd.merge(active_emp[['Employee_Id', 'Full_Name', 'Department', 'Job_Role']], 
                       talent_stats, on='Employee_Id')
```

```python
# 4. Lọc "Gà nòi": Performance > 4.0 và sắp xếp theo độ gắn kết + giờ học
final_list = top_talents[top_talents['Performance_Rating'] > 4.0].sort_values(
    by=['Engagement_Index', 'Training_Hours'], 
    ascending=[False, False]
)
```

```python
# Hiển thị Top 10 người đứng đầu
display(final_list.head(10))
```

```python
#Khảo sát Những cửa hàng có nhân viên "hạnh phúc" (Satisfaction cao) thì doanh số có thực sự bùng nổ hơn không?
df_business = all_sheets['business_outcomes']
```

```python
df_business.head()
```

```python
# 1. Tính TỔNG doanh thu của từng cửa hàng
store_revenue = df_business.groupby('Store_Id')['Sales_Actual'].sum().reset_index()
```

```python
# 2. Tính TRUNG BÌNH độ hài lòng của từng cửa hàng (Gộp từ bảng Employee và Performance)
emp_perf = pd.merge(df_performance, df_employee[['Employee_Id', 'Store_Id']], on='Employee_Id', how='inner')
store_happiness = emp_perf.groupby('Store_Id')['Employee_Satisfaction'].mean().reset_index()
```

```python
# 3. Nối 2 bảng kết quả lại với nhau bằng Store_Id
store_impact = pd.merge(store_revenue, store_happiness, on='Store_Id', how='inner')
```

```python
# 4. Hiển thị top 10
top_stores = store_impact.sort_values(by='Sales_Actual', ascending=False).head(10)
top_stores['Sales_Actual'] = top_stores['Sales_Actual'].apply(lambda x: f"{x:,.0f}")
top_stores['Employee_Satisfaction'] = top_stores['Employee_Satisfaction'].round(2)
```
